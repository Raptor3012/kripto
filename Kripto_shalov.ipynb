Задание 3.  
Дана строка закодированная в 16-виде. Данную строку получили путем операции XOR некоторого текста с одним символом. Расшифруйте это сообщение.   
Пример:  
Входные данные:  
19367831362e3d2b2c353d362c783136783336372f343d3c3f3d7839342f39212b782839212b782c303d783a3d2b2c7831362c3d2a3d2b2c  

```python
input_string = "19367831362e3d2b2c353d362c783136783336372f343d3c3f3d7839342f39212b782839212b782c303d783a3d2b2c7831362c3d2a3d2b2c"
```

```python
def split_string_bystep(input_string, step):    
    bigram_count = {}
    bigram_list= []
    for i in range(0, len(input_string), step):
        bigram_list.append(input_string[i:i+step])
        if input_string[i:i+step] not in bigram_count.keys():
            bigram_count[input_string[i:i+step]] = 1
        else:
            bigram_count[input_string[i:i+step]] += 1
    return bigram_count, bigram_list
```

```python
def max_frequency(bigram_count):
    max_element = 0
    for key,value in bigram_count.items():
        if value > max_element:
            max_element = value
    max_freq = []

    for key,value in bigram_count.items():
        if value == max_element:
            max_freq.append(key)
    return max_freq
```

```python
bigram_count, bigram_list = split_string_bystep(input_string, 2)
max_friq_bigram = max_frequency(bigram_count)
```

```python
char_e = "e".encode('ascii').hex()
keys = [hex(int(char_e, 16) ^ int(a, 16)) for a in max_friq_bigram]
```

```python
output = []
for key in keys:
    result = []
    for i in range(len(bigram_list)):
        result.append(chr(int(bigram_list[i], 16) ^ int(key, 16)))
    result = ''.join(result)
    output.append(result)
```

```python
result = [re.sub("[^ a-zA-ZА-Яа-я-.?!)(,:;]","",i) for i in output]
for i in result:
    if len(i) >= len(input_string)/2 * 0.95:
        print(i)
```
